```python
import torch
import torch.nn.functional as F

Tensor = torch.Tensor

# ---------- 1) Household Utility (CES in c, separable disutility of hours) ----------
class HouseholdUtility:
    def __init__(self, alpha1=0.65, alpha2=0.05, alpha3=0.3, rho=-2.0, mu=0.0545):
        self.alpha = torch.tensor([alpha1, alpha2, alpha3], dtype=torch.double)  # weights for c
        self.rho = rho   # CES parameter (rho=0 -> Cobb-Douglas in logsの極限)
        self.mu = mu     # disutility curvature on hours

    def u(self, c: Tensor, h: Tensor, N: Tensor, lam: float) -> Tensor:
        """
        c: (3,)  consumption per sector [c1,c2,c3]
        h: (3,)  hours per sector      [h1,h2,h3]
        N: (3,)  employment shares     [N1,N2,N3]  (使い方は任意だが、重み付けに利用)
        lam: scalar weight on disutility
        """
        c = c.clamp_min(1e-8)
        h = h.clamp_min(0.0)
        # CES aggregator over c1,c2,c3
        if abs(self.rho) < 1e-12:
            # rho -> 0 limit ≈ Cobb-Douglas in logs: sum(alpha_i * log c_i)
            util_c = torch.sum(self.alpha * torch.log(c))
        else:
            util_c = torch.sum(self.alpha * (c ** self.rho)) ** (1.0 / self.rho)

        # disutility of hours (weighted by employment N)
        disutil_h = lam * torch.sum(N * (h ** (1.0 + self.mu)) / (1.0 + self.mu))
        return util_c - disutil_h


# ---------- 2) Matching ----------
class Matching:
    def __init__(self, s=0.5, phi=0.47):
        self.s = torch.tensor(s, dtype=torch.double)
        self.phi = torch.tensor(phi, dtype=torch.double)

    def e(self, m: Tensor, u: Tensor) -> Tensor:
        """
        m: (3,) vacancies/postings per sector  [m1,m2,m3]
        u: (3,) unemployed searchers per sector [u1,u2,u3]
        returns hires per sector e_i = s * m_i^phi * u_i^(1-phi)
        """
        m = m.clamp_min(1e-12)
        u = u.clamp_min(1e-12)
        return self.s * (m ** self.phi) * (u ** (1.0 - self.phi))


# ---------- 3) Labor Dynamics ----------
class LaborDynamics:
    def __init__(self, r=0.0125, matching: Matching=None):
        self.r = torch.tensor(r, dtype=torch.double)  # separation rate
        self.matching = matching or Matching()

    def N_next(self, N: Tensor, m: Tensor, u: Tensor) -> Tensor:
        """
        N_{i,t+1} = (1 - r) * N_{i,t} + e_i(m_i, u_i)
        """
        e = self.matching.e(m, u)
        return N*((1.0 - self.r)+ e)


# ---------- 4) Production ----------
class Production:
    def __init__(self):
        pass

    def x(self, N: Tensor, z: Tensor, h: Tensor, m: Tensor) -> Tensor:
        """
        x_i = z_i * N_i * (h_i - m_i)   # 例：m_i は募集・訓練等で時間を食うとみなす
        """
        return z * N * (h - m)  # 要件に応じて仕様変更OK（ここは雛形）


# ---------- 5) Planner: choose {m_t, h_t} to max Σ β^t u(c_t,h_t; N_t) ----------
import torch
import torch.nn.functional as F

class PTPlanner:
    def __init__(self, T=50, beta=0.96, lam=1.0):
        self.T = T
        self.beta = torch.tensor(beta, dtype=torch.double)
        self.lam = lam

        self.util = HouseholdUtility()
        self.match = Matching()
        self.ldyn = LaborDynamics(matching=self.match)
        self.prod = Production()

        self.N0 = torch.tensor([0.3, 0.3, 0.3], dtype=torch.double)  # 合計<=1を想定
        self.z  = torch.tensor([1.0, 1.0, 1.0], dtype=torch.double)

    # ←← この simulate をクラス内に入れてください（インプレース更新なし）
    def simulate(self, M_free: torch.Tensor, H_free: torch.Tensor):
        """
        決定変数（非負制約は softplus で表現）:
          M_free: (T,3) → m_t >= 0
          H_free: (T,3) → h_t >= 0
        制約（雇用動学や資源制約）はこの中で前進シミュレーションとして満たす。
        """
        device = M_free.device
        m_t = F.softplus(M_free)  # (T,3)
        h_t = F.softplus(H_free)  # (T,3)

        N_list   = [self.N0.to(device)]
        u_list   = []
        x_list   = []
        c_list   = []
        u_agg_ls = []

        for t in range(self.T):
            N_prev = N_list[-1]                # (3,)
            u_total = 1.0 - N_prev.sum()
            share   = (N_prev / N_prev.sum().clamp_min(1e-12)).clamp_min(0.0)
            u_vec   = u_total * share          # (3,)

            x_t = self.prod.x(N_prev, self.z.to(device), h_t[t], m_t[t])  # (3,)
            c_t = x_t                              # 簡略：生産=消費

            u_t = self.util.u(c_t, h_t[t], N_prev, lam=self.lam)
            N_next = self.ldyn.N_next(N_prev, m_t[t], u_vec)

            u_list.append(u_vec)
            x_list.append(x_t)
            c_list.append(c_t)
            u_agg_ls.append(u_t)
            N_list.append(N_next)

        N     = torch.stack(N_list,   dim=0)   # (T+1,3)
        u_pool= torch.stack(u_list,   dim=0)   # (T,3)
        x_out = torch.stack(x_list,   dim=0)   # (T,3)
        c     = torch.stack(c_list,   dim=0)   # (T,3)
        u_agg = torch.stack(u_agg_ls, dim=0)   # (T,)

        betas = self.beta ** torch.arange(self.T, dtype=torch.double, device=device)
        total_util = torch.sum(u_agg * betas)

        return total_util, {"N": N, "m": m_t, "h": h_t, "c": c, "x": x_out, "u_per_t": u_agg, "u_pool": u_pool}

    # ←← これが新規：クラス内の optimize（Adamで効用最大化）
    def optimize(self, steps=1000, lr=1e-2, device=None):
        device = device or ("cuda" if torch.cuda.is_available() else "cpu")
        # 決定変数（自由パラメータ）を作って勾配対象に
        M_free = torch.zeros((self.T, 3), dtype=torch.double, requires_grad=True, device=device)
        H_free = torch.zeros((self.T, 3), dtype=torch.double, requires_grad=True, device=device)

        # 参照テンソルも同じ device に
        self.N0 = self.N0.to(device)
        self.z  = self.z.to(device)

        opt = torch.optim.Adam([M_free, H_free], lr=lr)

        # （デバッグしたい場合は次の行のコメントを外す）
        # torch.autograd.set_detect_anomaly(True)

        for s in range(1, steps+1):
            opt.zero_grad()
            total_util, _ = self.simulate(M_free, H_free)  # 制約付き前進→目的
            loss = -total_util                             # 最大化→最小化に反転
            loss.backward()
            opt.step()
            if s % 100 == 0:
                print(f"[step {s}] utility={float(total_util):.6f}")

        return M_free.detach(), H_free.detach()
```

```python
if __name__ == "__main__":
    planner = PTPlanner(T=50, beta=0.99, lam=1.0)
    M_opt, H_opt = planner.optimize(steps=500, lr=5e-3)
    total_util, paths = planner.simulate(M_opt, H_opt)
    print("Final utility:", float(total_util))
```

```python
# ===== αショックIRF（失業率と総生産）: 完全版 =====
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch

def set_alpha(planner: PTPlanner, alpha_vec):
    """planner.util.alpha を差し替え（正規化付き）"""
    a = torch.as_tensor(alpha_vec, dtype=torch.double)
    a = a.clamp_min(1e-12)
    a = a / a.sum()
    planner.util.alpha = a
    return planner

def _solve_paths(planner: PTPlanner, steps=600, lr=5e-3, device=None):
    """最適化→パス取得"""
    M_opt, H_opt = planner.optimize(steps=steps, lr=lr, device=device)
    util, paths = planner.simulate(M_opt, H_opt)
    return float(util), paths

def _agg_series(paths):
    """
    失業率: u_t = 1 - Σ_i N_{i,t}（期tの雇用から）
    総生産: y_t = Σ_i x_{i,t}
    """
    N = paths["N"].detach().cpu().numpy()        # (T+1,3)
    X = paths["x"].detach().cpu().numpy()        # (T,3)
    u_rate = 1.0 - N[:-1].sum(axis=1)            # (T,)
    y_total = X.sum(axis=1)                      # (T,)
    return u_rate, y_total

def irf_alpha(planner_base: PTPlanner,
              alpha_new,
              permanent=True,
              steps_base=600, steps_shock=600, lr=5e-3, device=None,
              outdir="outputs", prefix="irf_alpha"):
    """
    αを変更したときのIRF（失業率・総生産）を作成。
    permanent=True: 恒久ショック（全期間 α_new）
    """
    # 1) ベースライン
    base_pln = copy.deepcopy(planner_base)
    util0, paths0 = _solve_paths(base_pln, steps=steps_base, lr=lr, device=device)
    u0, y0 = _agg_series(paths0)  # (T,), (T,)

    # 2) ショック後
    shock_pln = copy.deepcopy(planner_base)
    set_alpha(shock_pln, alpha_new)
    util1, paths1 = _solve_paths(shock_pln, steps=steps_shock, lr=lr, device=device)
    u1, y1 = _agg_series(paths1)

    # 3) IRF（差分と%変化を両方出す）
    eps = 1e-12
    u_irf_abs = (u1 - u0)                    # 失業率の差分
    y_irf_abs = (y1 - y0)                    # 生産の差分
    u_irf_pct = 100.0 * (u1 - u0) / (np.abs(u0) + eps)
    y_irf_pct = 100.0 * (y1 - y0) / (np.abs(y0) + eps)

    # 4) 可視化
    tt = np.arange(len(u0))
    fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
    ax[0].plot(tt, u_irf_pct); ax[0].axhline(0, lw=0.8)
    ax[0].set_ylabel("Unemployment IRF (%Δ)")
    ax[0].set_title("IRF to Alpha shock (preference weights)")
    ax[1].plot(tt, y_irf_pct); ax[1].axhline(0, lw=0.8)
    ax[1].set_ylabel("Output IRF (%Δ)")
    ax[1].set_xlabel("t")
    plt.tight_layout()
    plt.show()

    # 5) CSV 保存
    import os
    os.makedirs(outdir, exist_ok=True)
    pd.DataFrame({"t": tt, "u_base": u0, "u_shock": u1,
                  "u_irf_abs": u_irf_abs, "u_irf_pct": u_irf_pct}
                 ).to_csv(f"{outdir}/{prefix}_unemp.csv", index=False)
    pd.DataFrame({"t": tt, "y_base": y0, "y_shock": y1,
                  "y_irf_abs": y_irf_abs, "y_irf_pct": y_irf_pct}
                 ).to_csv(f"{outdir}/{prefix}_output.csv", index=False)

    # 元パスを保存（任意）
    def save_full(paths, tag):
        pd.DataFrame(paths["N"].detach().cpu().numpy()).to_csv(f"{outdir}/{prefix}_{tag}_N.csv", index=False)
        pd.DataFrame(paths["x"].detach().cpu().numpy()).to_csv(f"{outdir}/{prefix}_{tag}_X.csv", index=False)
        pd.DataFrame(paths["c"].detach().cpu().numpy()).to_csv(f"{outdir}/{prefix}_{tag}_C.csv", index=False)
    save_full(paths0, "base")
    save_full(paths1, "shock")

    # 6) 戻り値（可視化に再利用できるよう %IRF を返す）
    return (tt, u_irf_pct, y_irf_pct), (paths0, paths1)
```

```python
planner = PTPlanner(T=50, beta=0.99, lam=1.0)
(tt, u_irf, y_irf), (base_paths, shock_paths) = irf_alpha(
    planner_base=planner,
    alpha_new=[0.3, 0.3, 0.4],  # 例：αを変更
    permanent=True,
    steps_base=200, steps_shock=200, lr=5e-3,
    outdir="outputs", prefix="alpha_perm"
)

# 追加プロット例（任意）
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(tt, u_irf, label="Unemployment IRF (%Δ)")
plt.plot(tt, y_irf, label="Output IRF (%Δ)")
plt.axhline(0, lw=0.8)
plt.legend(); plt.xlabel("t"); plt.title("Alpha Shock IRF"); plt.show()
```
